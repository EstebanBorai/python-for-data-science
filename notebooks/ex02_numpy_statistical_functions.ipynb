```python
import numpy as np

jeff_salary = [2700, 3000, 3000]
nick_salary = [2600, 2800, 2800]
tom_salary = [2300, 2500, 2500]

base_salary = np.array([jeff_salary, nick_salary, tom_salary])

print(base_salary)
```

```python
jeff_bonus = [500, 400, 400]
nick_bonus = [600, 300, 400]
tom_bonus = [200, 500, 400]

bonus = np.array([jeff_bonus, nick_bonus, tom_bonus])

print(bonus)
```

```python
salary_bonus = base_salary + bonus
print(salary_bonus)
```

```python
highest_salary = salary_bonus.max()
print(f'Highest Salary is {highest_salary}')
```

Finds the highest salary per month using `np.amax`.
In NumPy Arrays theres 2 axes (due to the bi-dimensional array) available `0` to represent rows and `1` to represent columns.



Finds the average of the maximum amount paid each month to all employees

```python
print('Salaries')
print(salary_bonus)
print('')

highest_salary_per_employee = np.amax(salary_bonus, axis = 1)
print(f'Highest Salary Paid per Employee: {highest_salary_per_employee}')

average_highest_salary = np.average(highest_salary_per_employee)
print(f'Average from the Highest Salary Paid per Employee: {average_highest_salary}')
```
